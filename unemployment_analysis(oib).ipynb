```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
```

```python
X = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")
```

```python
X
```

```python
print(X.isnull().sum())
```

```python

X.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate",
               "Estimated Employed",
               "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]
```

```python
X.describe()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12, 12))
sns.heatmap(X.corr(),cmap="crest")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Employment Rate")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=X)
plt.show()
```

This is the analysis of unemployment rate in India dataset.
By the analysis of these graphs,it tells us that unemployment rate in south region is more with count approximately 20.5
